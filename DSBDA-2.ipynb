```python
import pandas as pd

df['Reading_score'] = pd.to_numeric(df['Reading_score'], errors='coerce')

sorted_rsscore = sorted(df['Reading_score'].dropna())

print(sorted_rsscore)
```

```python
q1 = np.percentile(sorted_rsscore, 25)
q2 = np.percentile(sorted_rsscore, 75)
print(q1, q2)
```

```python
IQR = q2 - q1
```

```python
lower = q1-(1.5*IQR)
upper = q2+(1.5*IQR)
print(lower, upper)
```

```python
res = []
for i in sorted_rsscore:
    if(i<lower or i>upper):
        res.append(i)
print(res)
```
